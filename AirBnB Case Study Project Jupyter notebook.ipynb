```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb.tail()
```

```python
# How many neighborhood groups are available and which shows up the most?
air_neighbourhood=air_bnb.groupby('neighbourhood_group', as_index=False).count()[['id','neighbourhood_group','price']]
air_neighbourhood.head()
# There are 5 neighborhood groups, and Manhattan shows up with the most, Brooklyn close behind
```

```python
# Are private rooms the most popular in manhattan?
air_man_private = air_bnb.groupby(['neighbourhood_group', 'room_type']).count()
air_man_private
# Private rooms are 2nd most popular in manhattan after entire home/apt
```

```python
# Which hosts are the busiest and based on their reviews?
air_hosts = air_bnb.groupby('host_id').sum()[['id','number_of_reviews']]
air_hosts.sort_values(by=['number_of_reviews'],ascending=False).head()
# Host id 37312959 is busiest and 344035 is second busiest
```

```python
#Which neighorhood group has the highest average price?
air_group_price = air_bnb.groupby('neighbourhood_group').mean().round(decimals=2)[['id','price']]
air_group_price.sort_values(by=['price'],ascending=False).head()
# Manhattan has the highest average price
```

```python
# Which neighbor hood group has the highest total price?
air_group_price = air_bnb.groupby('neighbourhood_group').sum()[['id','price']]
air_group_price.sort_values(by=['price'],ascending=False).head()
#Is this what this question is asking? If so, Manhattan
```

```python
#Which top 5 hosts have the highest total price?
air_host_price = air_bnb.groupby('host_id').sum()[['id','price']]
air_host_price.sort_values(by=['price'],ascending=False).head()
#Host IDs 219517861, 107434423, 156158778, 205031545, and 30283594 have the highest total price
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_zero_avail = air_bnb[air_bnb['availability_365'] == 0]
air_zero_reviewed = air_zero_avail[air_zero_avail['number_of_reviews'] >= 100]
air_zero_reviewed
# There are 162 people with a review count of 100 or more with no availability
```

```python
# What host has the highest total of prices and where are they located?
air_host_price = air_bnb.groupby('host_id').sum()[['id','price',]]
air_host_price.sort_values(by=['price'],ascending=False).head()
# Host ID 219517861 has the highest total of prices
air_host_id_loc = air_bnb[air_bnb['host_id']==219517861]
air_host_id_loc
# They are located in Manhattan
```

```python
# When did Danielle from Queens last receive a review?
air_danielle = air_bnb[air_bnb['host_name']=='Danielle']
air_danielle_queens = air_danielle[air_danielle['neighbourhood_group']=='Queens']
air_danielle_queens[['host_name','host_id','last_review']]
# There are multiple Danielle's from Queens, but the lastest review was on 2019-07-08
```
